```python
# Solution for Number 1
import pandas as pd
import numpy as np

# LOAD DATA
df = pd.read_csv(r"C:\Users\Sandrex\OneDrive\Desktop\comprog_lab\Lab6_NumPyAndPandas_Dala\Most Popular Programming Languages.csv")
df.columns = df.columns.str.strip()
df["Month"] = pd.to_datetime(df["Month"])

# LANGUAGE SELECTION
language_cols = sorted([c for c in df.columns if c != "Month"])

student_id = "TUPM-25-3410"
idx = int(student_id[-3:]) % len(language_cols)
lang = language_cols[idx]

print("Assigned Language:", lang)

# DATA PREP
data = df[["Month", lang]].copy()
data.rename(columns={"Month": "Date", lang: "Popularity"}, inplace=True)

# GROWTH + ROLLING STATS
data["Growth_Rate"] = data["Popularity"].pct_change() * 100
data["Moving_Avg"] = data["Popularity"].rolling(3).mean()
data["Moving_STD"] = data["Popularity"].rolling(3).std()

# PHASE CLASSIFICATION
data["Phase"] = np.select(
    [data["Growth_Rate"] > 5, data["Growth_Rate"] < -5],
    ["Growth", "Decline"],
    default="Stable"
)

# STATS
initial, final = data["Popularity"].iloc[0], data["Popularity"].iloc[-1]
overall_growth = ((final - initial) / initial) * 100

summary = data["Popularity"].describe()
phase_counts = data["Phase"].value_counts()

# OUTPUT TABLE
result = data[["Date","Popularity","Growth_Rate","Moving_Avg","Moving_STD","Phase"]]

print("\nDATA SAMPLE:\n", result.head())
print("\nSUMMARY:\n", summary)
print("\nPHASE COUNTS:\n", phase_counts)
print("\nMETRICS:")
print("Initial:", initial)
print("Final:", final)
print("Overall Growth %:", overall_growth)
```

```python
# Solution for Number 2
# LOAD
df = pd.read_csv(r"C:\Users\Sandrex\OneDrive\Desktop\comprog_lab\Lab6_NumPyAndPandas_Dala\Most Popular Programming Languages.csv")
df.columns = df.columns.str.strip()
df["Month"] = pd.to_datetime(df["Month"])

# SELECT LANGUAGE
langs = sorted([c for c in df.columns if c != "Month"])
i = int("TUPM-25-3410"[-3:]) % len(langs)
lang = langs[i]

data = df[["Month", lang]].rename(columns={"Month":"Date", lang:"Popularity"})

# GROWTH + SMOOTHING
data["Growth"] = data["Popularity"].pct_change() * 100
data["MA"] = data["Popularity"].rolling(6).mean()
data["STD"] = data["Popularity"].rolling(6).std()

# STATS
m, s = np.mean(data["Growth"].dropna()), np.std(data["Growth"].dropna())

# LIFECYCLE
data["Phase"] = np.select(
    [
        (data["Growth"] > 0) & (data["Growth"] < m),
        (data["Growth"] > m),
        (np.abs(data["Growth"]) <= 1),
        (data["Growth"] < -s)
    ],
    ["Intro","Growth","Mature","Decline"],
    default="Stable"
)

# OUTPUT
print("LANG:", lang)
print(data.head())
print("\nCOUNT:\n", data["Phase"].value_counts())
print("\nDOMINANT:", data["Phase"].value_counts().idxmax())
```

```python
# Solution for Number 3
# LOAD DATA
df = pd.read_csv(r"C:\Users\Sandrex\OneDrive\Desktop\comprog_lab\Lab6_NumPyAndPandas_Dala\Most Popular Programming Languages.csv")

df.columns = df.columns.str.strip()
df["Month"] = pd.to_datetime(df["Month"])

# LANGUAGE SELECTION
langs = sorted([c for c in df.columns if c != "Month"])

student_id = "TUPM-25-3410"
i = int(student_id[-3:]) % len(langs)

lang1 = langs[i]
lang2 = langs[(i + 1) % len(langs)]

print("Language A:", lang1)
print("Language B:", lang2)

# DATA
a = df[lang1].values
b = df[lang2].values

# STATS
mean_a, mean_b = np.mean(a), np.mean(b)
std_a, std_b = np.std(a), np.std(b)

# CORRELATION
corr = np.corrcoef(a, b)[0, 1]

# DOMINANCE
dom_ratio = np.sum(a > b) / len(df) * 100

# RPI
rpi = (mean_a - mean_b) / (std_a + std_b)

# CROSSOVERS
crossovers = np.sum((a[:-1] < b[:-1]) & (a[1:] > b[1:]))

# SUMMARY
summary = pd.DataFrame([{
    "Mean_A": mean_a,
    "Mean_B": mean_b,
    "Std_A": std_a,
    "Std_B": std_b,
    "Correlation": corr,
    "Dominance_%": dom_ratio,
    "RPI": rpi
}])

print("\nSUMMARY:\n", summary)
print("\nCrossovers:", crossovers)
```
